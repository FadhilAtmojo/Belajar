Diberikan Class `BinaryTree` dan Class `BinaryTreeNode` seperti di bawah. Tambahkan fungsi `insert` untuk menyisipkan node pada Tree sebagi child dari suatu node yang sudah ada.

Fungsi ini memiliki 3 parameter:
- new_data : data node yang akan ditambahkan
- parent_data : node yang sudah ada yang akan menjadi parent dari node yang akan ditambahkan
- "left"/"right" : posisi node yang akan di-insert ada di kiri atau di kanan dari parentnya

Jika posisi penyisipan telah terisi, outputkan `Posisi sudah terisi`. Jika parent node tidak ditemukan, outputkan `Parent tidak ditemukan`

Perhatikan contoh cell di bawahnya. Jika fungsi insert yang dibuat benar, maka script pada cell tersebut akan memberikan output yang sesuai

```python
class BinaryTreeNode:
  def __init__(self, data):
    self.data  = data    # root node
    self.left  = None    # left child
    self.right = None    # right child

class BinaryTree:
    def __init__(self, root):
        self.root = root #root node

    def display(self):
        self.preorder(self.root)

    def findnode(self, x):
        return self.preorderfind(self.root, x)
    
    def preorder(self, node_now):
        if node_now!=None:
            print(node_now.data)
            self.preorder(node_now.left)
            self.preorder(node_now.right)

    # TODO
    def insert(self, new_data, parent_data, pointer):
      if self.preorderfind(self.root, parent_data):
        node_baru = BinaryTreeNode(new_data)
        if pointer == "left":
          x = self.preorderfind(self.root, parent_data)
          if x.left == None:
            x.left = node_baru
          else:
            print("Posisi sudah terisi!")
        elif pointer == "right":
          x = self.preorderfind(self.root, parent_data)
          if x.right == None:
            x.right = node_baru
          else:
            print("Posisi sudah terisi!")
      else:
        print("parent tidak ditemukan!")

        

    def preorderfind(self, node_now, x):
        if node_now!=None:
            if node_now.data==x: # bandingkan dengan nilai di node ini
                return node_now # yang dikembalikan nodenya

            # kalau belum ketemu  ke kiri, cari nilainya
            hasil_left = self.preorderfind(node_now.left, x)
            if hasil_left:
                return hasil_left

            # kalau belum ketemu  ke kanan, cari nilainya
            hasil_right = self.preorderfind(node_now.right, x) 
            if hasil_right:
                return hasil_right
        else:
            return None # kalau sampai ujung nggak ada hasilnya 0
```

```python
new_rootnode = BinaryTreeNode(3)

bt = BinaryTree(new_rootnode) 
bt.display()
# visualisasi seharusnya
#        [3]

print("---------------")
bt.insert(11, 5, "right") # tidak ada node 5, muncul pesan error

print("---------------")
bt.insert(11, 3, "right")
bt.insert(10, 3, "left")
bt.display()
# visualisasi seharusnya
#        [3]
#       /    \
#     [10]   [11]

print("---------------")
bt.insert(24, 10, "left")
bt.display()
# visualisasi seharusnya
#        [3]
#       /    \
#     [10]   [11]
#    /    
#  [24]  

print("---------------")
bt.insert(15, 3, "left") # anak kiri dari 3 sudah ada, muncul pesan error
bt.display()
# visualisasi seharusnya
#        [3]
#       /    \
#     [10]   [11]
#    /    
#  [24]  
```
