```python
# Nomer 2
n, m = map(int, input().split())

left = []
right = []

for i in range(n):
    name, side = input().split()
    if side == "kiri":
        left.append(name)
    else:
        right.append(name)
print()
while len(left) > 0:
    print("kiri ke kanan:", end=" ")
    onboard = []
    while len(onboard) < m and len(left) > 0:
        onboard.append(left.pop(0))
    print(" ".join(onboard))
    
    if len(left) == 0 and len(right) == 0:
        break
    
    print("kanan ke kiri:", end=" ")
    onboard = []
    while len(onboard) < m and len(right) > 0:
        onboard.append(right.pop(0))
    print(" ".join(onboard))
    
```
